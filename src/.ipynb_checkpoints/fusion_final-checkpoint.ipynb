```python
import re
import pandas as pd
import numpy as np
import requests
import sys
```

```python
f=pd.read_csv('../data/fusion.txt',sep='\t')
```

```python
def get_seq(strs):
    # split by {} to get transcript name
    name = [", ".join(x.split()) for x in re.split(r'[{}]',strs) if x.strip()] 

    try:
        transcript_up_ID = name[1]
        transcript_down_ID = name[3]

        # get exon posision
        t_up_location = [", ".join(x.split()) for x in re.split('\+|\_',name[2]) if x.strip()]
        t_down_location = [", ".join(x.split()) for x in re.split('\-|\_|\.',name[4]) if x.strip()]

        transcript_up_start = 1
        if '+' in name[2]:  
            # with intron info
            transcript_up_end = int(t_up_location[1])
            transcript_up_intron = int(t_up_location[2])
            if 'ins' in name[2]:
                ins = t_up_location[3].replace('ins','')
        else:
            transcript_up_end = int(t_up_location[1])
            transcript_up_intron = 0
            if 'ins' in name[2]:
                ins = t_up_location[2].replace('ins','')
                
        if '-' in name[4]:
            # with intron info
            transcript_down_start = int(t_down_location[1])
            transcript_down_intron = int(t_down_location[2])
            transcript_down_end = int(t_down_location[3])
        else:
            transcript_down_start = int(t_down_location[1])
            transcript_down_intron = 0
            transcript_down_end = int(t_down_location[2])

        try:
            # use api to get sequence 
            if 'ENST' in transcript_up_ID: 
                # ensembl ID
                server = "https://grch37.rest.ensembl.org"
                t_up = "/sequence/id/" + transcript_up_ID + "?type=cdna"
                r = requests.get(server + t_up, headers = { "Content-Type" : "text/plain"}) 
                if not r.ok:
                    r.raise_for_status()
                    sys.exit()
                seq_up = r.text
            else:
                # refseq ID
                r = requests.get('https://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi?db=nucleotide&id='+transcript_up_ID+'&rettype=FASTA').text                
                seq_up = ''.join(r.split('\n')[1:-2])
                
            l.append(strs)
            if transcript_up_end >= 100:
                l.append(seq_up[(transcript_up_end - 100):(transcript_up_end)])
            else:
                l.append(seq_up[0:(transcript_up_end)])
            try:
                l.append(ins)
            except:
                pass
            
            if 'ENST' in transcript_down_ID:
                t_down = "/sequence/id/" + transcript_down_ID + "?type=cdna"
                r = requests.get(server + t_down, headers = { "Content-Type" : "text/plain"}) 
                if not r.ok:
                    r.raise_for_status()
                    sys.exit()  
                seq_down = r.text
            else:
                r = requests.get('https://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi?db=nucleotide&id='+transcript_down_ID+'&rettype=FASTA').text                
                seq_down = ''.join(r.split('\n')[1:-2])
            
            l.append(seq_down[(transcript_down_start - 1):(transcript_down_start -1+100)])
            
        except:
            print(strs,'failed')
        return(l)
    except:
        print(name,'failed')
```

```python

l = []
for idx,val in f.iterrows():
    strs = val['translocation name'].replace("+?",'').replace("-?",'')
    # remove unknonw intron break point
    if '_o' not in strs and '(' not in strs:  
        print(strs,'ongoing')
        # keep track of process
        l = get_seq(strs)
    else:
        print(strs)
```

```python
pd.DataFrame(l).to_csv('fusion_seq_cleanup_grch37.txt', sep='\t', index=False, header=False)
```

```python
# some position is incorrect on transcript, possibly due to different ID, need additional step 
# to check if the position is on the exon edge
from xml.etree import ElementTree
import time
```

```python
# build dic to store exon edge
refseq_exon_dic = dict()
for i in ['NM_000142','NM_002944','NM_004304','NM_020975', 'NM_003466.2','NM_005933.1','NM_015869.2','NM_024426.2']:
    print(i)
    time.sleep(3) # error without sleep
    l=[]
    req = requests.get('https://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi?db=nuccore&id='+i+'&retmode=xml&rettype=native')
    tree = ElementTree.fromstring(req.content)

    for start in tree.findall('.//Seq-interval_from'):
        l.append(int(start.text))
    for end in tree.findall('.//Seq-interval_to'):
        l.append(int(end.text))
    refseq_exon_dic[i] = l
```

```python
def check_refseq(strs):
    name = [", ".join(x.split()) for x in re.split(r'[{}]',strs) if x.strip()] 
    try:
        transcript_up_ID = name[1]
        transcript_down_ID = name[3]
        t_up_location = [", ".join(x.split()) for x in re.split('\+|\_',name[2]) if x.strip()]
        t_down_location = [", ".join(x.split()) for x in re.split('\-|\_|\.',name[4]) if x.strip()]
        transcript_up_end = int(t_up_location[1])
        transcript_down_start = int(t_down_location[1])

        try:
            if 'NM' in transcript_up_ID: 
                if transcript_up_end-1 not in refseq_exon_dic[transcript_up_ID]:
                    wrong_exon.append(strs)
                else:
                    print('right',strs)           
            if 'NM' in transcript_down_ID:
                if transcript_down_start-1 not in refseq_exon_dic[transcript_down_ID]:
                    wrong_exon.append(strs)
                else:
                    print('right',strs)
                    
        except:
            print(strs,'failed')
        return(wrong_exon)
    except:
        print(name,'failed')
```

```python
wrong_exon = []
for idx,val in f.iterrows():
    strs = val['translocation name'].replace("+?",'').replace("-?",'')
    if '_o' not in strs and '(' not in strs:  
        l = check_refseq(strs)
    else:
        pass
```
